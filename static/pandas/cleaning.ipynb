```python
import pandas as pd
import os
```

```python
file_path = "../data/overalldata_cleaned.csv"
file_one_df = pd.read_csv(file_path, encoding="ISO-8859-1")
```

```python
file_one_df.head()
```

```python
##Replace date into just years
only_years = []
for date in file_one_df["Date"]:
    only_years.append(date.split("/")[2])

file_one_df["Date"] = only_years
```

```python
##Groupby functions to get total coun

##casualties table
casualties = file_one_df.groupby(['Date']).sum()
```

```python
## Total number of shootings per year
Total_Shoots = pd.DataFrame(file_one_df['Date'].value_counts())
Total_Shoots = Total_Shoots.sort_index()
```

```python
## Total number of shootings per year by Area Type
urban = pd.DataFrame(file_one_df.loc[(file_one_df["AreaType"]== "urban") | (file_one_df["AreaType"]== "Urban") ])
urban_shoots = pd.DataFrame(urban['Date'].value_counts())
urban_shoots = urban_shoots.sort_index()

urban_shoots.reset_index(inplace=True)
urban_shoots.rename(columns={'index': 'Year'}, inplace=True)

urban_array = []
for date in urban_shoots['Date']:
    urban_array.append("urban")

urban_shoots['AreaType']=urban_array
urban_shoots
```

```python
rural = pd.DataFrame(file_one_df.loc[(file_one_df["AreaType"]== "rural") | (file_one_df["AreaType"]== "Rural") ])
rural_shoots = pd.DataFrame(rural['Date'].value_counts())
rural_shoots = rural_shoots.sort_index()

rural_shoots.reset_index(inplace=True)
rural_shoots.rename(columns={'index': 'Year'}, inplace=True)

rural_array =[]
for date in rural_shoots['Date']:
    rural_array.append("rural")
    
rural_shoots['AreaType']=rural_array
rural_shoots
```

```python
suburban = pd.DataFrame(file_one_df.loc[(file_one_df["AreaType"]== "suburban") | (file_one_df["AreaType"]== "Suburban") ])
suburban_shoots = pd.DataFrame(suburban['Date'].value_counts())
suburban_shoots = suburban_shoots.sort_index()

suburban_shoots.reset_index(inplace=True)
suburban_shoots.rename(columns={'index': 'Year'}, inplace=True)

suburban_array = []
for date in suburban_shoots['Date']:
    suburban_array.append("suburban")

suburban_shoots['AreaType']=suburban_array
suburban_shoots
```

```python
#Cleaning up the csvs to something usable
Total_Shoots.reset_index(inplace=True)
Total_Shoots.rename(columns={'index': 'Year'}, inplace=True)
```

```python
#merge the urban,rural,suburban dataframes
merged1 = urban_shoots.merge(rural_shoots, how="outer")
merged2 = merged1.merge(suburban_shoots, how = "outer")
```

```python
Total_Shoots.to_csv("../data/Total_Shootings.csv", index=False)
merged2.to_csv("../data/AreaType_Shootings.csv", index=False)
```
